<h3><b>Popularity Based Recommender<b>

-- Now we'll merge the two dataframes users and books 

-- Then see the no. of ratings by the users for each book 

-- With this Info we can decide which book has genuine ratings 

```python
ratings_books=ratings.merge(books,on='ISBN')
ratings_books.head()
```

```python
ratings_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
```
